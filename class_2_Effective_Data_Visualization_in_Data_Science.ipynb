## CODING TASK #4. PLOT HISTOGRAMS USING MATPLOTLIB
วาดฮิสโตแกรมโดยใช้ Matplotlib

```python
# A histogram represents data using bars with various heights
# Each bar groups numbers into specific ranges
# Taller bars show that more data falls within that specific range

# ฮิสโตแกรมแสดงข้อมูลโดยใช้แท่งที่มีความสูงต่างกัน
# แต่ละแท่งจะจัดกลุ่มตัวเลขเข้าไปในช่วงที่กำหนด
# แท่งที่สูงกว่าบ่งบอกว่ามีข้อมูลมากขึ้นอยู่ในช่วงนั้น

daily_return_df = pd.read_csv('crypto_daily_returns.csv')
daily_return_df
```

```python
mu = round(daily_return_df['BTC'].mean(), 2) # mean of distribution # ค่ากลางของการแจกแจง
sigma = round(daily_return_df['BTC'].std(), 2) # standard deviation of distribution # ส่วนเบี่ยงเบนมาตรฐานของการแจกแจง

num_bins = 40

# Plot the histogram of the returns
# วาดแผนภูมิฮิสโทแกรม
plt.figure(figsize = (20, 15))

daily_return_df['BTC'].plot.hist(bins = num_bins, color = 'b', alpha = 0.7, figsize = (12, 6))

plt.grid()
plt.ylabel('Probability')
plt.title('Histogram: mu=' + str(mu) + ', sigma=' +str(sigma));
```
